## Problem statement:
### Perform Clustering(Hierarchical, Kmeans & DBSCAN) for the crime data and identify the number of clusters formed and draw inferences.
## Dataset: crime_data.csv


```python
# Importing libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
# Loading Dataset
data = pd.read_csv('crime_data.csv')
data
```

## Data Processing and EDA

```python
data = data.rename({'Unnamed: 0': 'State'}, axis = 1)
data.head()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data.describe()
```

## Agglomerative Clustering

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Normalizing dataset

scalar = MinMaxScaler()
scalar_df = scalar.fit_transform(data.iloc[:,1:])
print(scalar_df)
```

```python
# Cerating dendrograms
plt.figure(figsize=(10,7))
dendogram = sch.dendrogram(sch.linkage(scalar_df,'complete'))
```

```python
# Creating clusters
H_clusters = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
H_clusters
```

```python
# using data normalized by MinMaxscalar
y = pd.DataFrame(H_clusters.fit_predict(scalar_df), columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
data['clustersid_HC'] = H_clusters.labels_
data
```

```python
data.groupby('clustersid_HC').agg(['mean']).reset_index()
```

```python
# Plotting barplot using groupby method to get visualization of how states in each cluster

fig, ax = plt.subplots(figsize=(10, 6))
data.groupby(['clustersid_HC']).count()['State'].plot(kind='bar')
plt.ylabel('States')
plt.title('Hierarchical Clustering',fontsize='large')
ax.set_xlabel('Clusters', fontsize='large')
ax.set_ylabel('States', fontsize='large')
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.show()
```

```python
# silhouette_score of AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
sil_score = silhouette_score(scalar_df, H_clusters.labels_)
sil_score
```

## KMEANS Clustering

```python
from sklearn.cluster import KMeans
```

### The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion:

```python
# using data normalized by MinMaxscalar
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scalar_df)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

```

### From above plot, optimum number of clusters can be selected equal to 4

```python
# Building cluster Algorithm
KM_clusters = KMeans(4, random_state=42)
KM_clusters.fit(scalar_df) 
```

```python
y = pd.DataFrame(KM_clusters.fit_predict(scalar_df),columns=['clusterid_Kmeans'])
y['clusterid_Kmeans'].value_counts() 
```

```python
#Assign clusters to the data set
data['clusterid_Kmeans'] = KM_clusters.labels_
data
```

```python
data.groupby('clusterid_Kmeans').agg(['mean']).reset_index()
```

```python
# Plotting barplot using groupby method to get visualization of how states in each cluster

fig, ax = plt.subplots(figsize=(10, 6))
data.groupby(['clusterid_Kmeans']).count()['State'].plot(kind='bar')
plt.ylabel('States')
plt.title('KMeans Clustering')
ax.set_xlabel('Clusters')
ax.set_ylabel('States')
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.show()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors  
```

```python
neigh = NearestNeighbors(n_neighbors=8)
nbrs = neigh.fit(scalar_df)
distances, indices = nbrs.kneighbors(scalar_df)  
```

```python
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)  
```

### We will try for different values of eps and min_samples

```python
EPS = [0.22, 0.24, 0.26,0.28, 0.30]

for n in EPS:
    dbscan = DBSCAN(eps=n, min_samples=4)
    dbscan.fit(scalar_df)
    y=pd.DataFrame(dbscan.fit_predict(scalar_df),columns=['clusterid_DBSCAN'])
    print(f'For eps = {n}','\n',y['clusterid_DBSCAN'].value_counts())
    
    # silhouette score
    sil_score= silhouette_score(scalar_df, dbscan.labels_)
    print(f'For eps = {n} silhouette score is ', sil_score)
```

### When we have value of epsilon = 0.28, we are getting 3 clusters silhouette score is more as compared to other dbscan models.
### -1 shows the noisy data points

```python
dbscan = DBSCAN(eps=0.28, min_samples=4)
dbscan.fit(scalar_df)
```

```python
data['clusterid_DBSCAN'] = dbscan.labels_
data.head()
```

```python
data.groupby('clusterid_DBSCAN').agg(['mean']).reset_index()
```

```python
# Plotting barplot using groupby method to get visualization of how many row no. in each cluster

fig, ax = plt.subplots(figsize=(10, 6))
data.groupby(['clusterid_DBSCAN']).count()['State'].plot(kind='bar')
plt.ylabel('ID Counts')
plt.title('DBSCAN Clustering')
ax.set_xlabel('Clusters')
ax.set_ylabel('States')
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.show()
```
